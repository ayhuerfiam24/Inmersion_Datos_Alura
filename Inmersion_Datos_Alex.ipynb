```python
#introducir una nueva columna
Valor=inmuebles.Valor.str.split(expand=True) #creo una variable
inmuebles["T/Moneda"]=Valor[0] #a la variable creada introduzco la nueva columna
inmuebles["Precio"]=Valor[1]
inmuebles.sample(3) #visualizo los primeros tres items
```

```python
inmuebles.info()
```

```python
inmuebles["Precio"]=inmuebles["Precio"].str.replace(".","",regex=True) #reeemplazar el punto de los valores de la columna
```

```python
inmuebles[["Precio","Barrio"]] #visualizar la colomna deseada
```

```python
#reducir la cifra precio
inmuebles["Precio_Millon"]=inmuebles.Precio.astype("float")/1000000
```

**Aula 3**

```python
inmuebles.head()
```

```python
inmuebles["Valor_m2_millon"]=inmuebles["Precio_Millon"]/inmuebles["Área"] #Calcular el valor del metro cuadrado
inmuebles.head()
```

```python
inmuebles.groupby("Barrio").mean() #Agrupar por una categoria
```

```python
datos_barrio=inmuebles.groupby("Barrio").sum()
datos_barrio
```

```python
datos_barrio["Valor_m2_barrio"]=datos_barrio["Precio_Millon"]/datos_barrio["Área"]
datos_barrio
```

```python
m2_barrio=dict(datos_barrio["Valor_m2_barrio"])
```

```python
inmuebles["Valor_m2_barrio"]=inmuebles["Barrio"]
inmuebles["Valor_m2_barrio"]=inmuebles["Valor_m2_barrio"].map(m2_barrio)
inmuebles.head(5)
```

```python
top_barrios=inmuebles["Barrio"].value_counts()[:10].index #los principales barrio para la venta
datos_barrio.reset_index(inplace=True)
datos_barrio
```

```python
datos_barrio.query("Barrio in @top_barrios") #devolver un dataframe de los 10 primeros barrios
```

```python
plt.figure(figsize=(10,8)) #graficar un barplot con el mayor numero de inmuebles en el dataset
ax=sns.barplot(x="Barrio",y="Valor_m2_barrio",data=datos_barrio.query("Barrio in @top_barrios")) #almacenamiento del grafico
ax.tick_params(axis="x", rotation=45) #ajustar los labels del eje x
```

```python
plt.figure(figsize=(10,8)) #graficar un barplot con el mayor numero de inmuebles en el dataset
ax=sns.boxplot(x="Barrio",y="Valor_m2_millon",data=inmuebles.query("Barrio in @top_barrios")) #almacenamiento del grafico
ax.tick_params(axis="x", rotation=45) #ajustar los labels del eje x
plt.show() #muestre el grafico como tal
```
